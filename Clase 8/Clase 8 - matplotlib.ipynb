# Clase 8 - matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x ** 2
```

```python
plt.plot(x, y, 'red', alpha=0.2)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.tittle("Gr√°fico Y en funcion de X")
```

## Sublopts en la misma celda

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'b-+')
```

## OO Matplotlib

```python
fig = plt.figure()
juan = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #izq, piso, ancho, altura
axes.plot(x, y)
axes.set_xlabel("Eje X")
axes.set_ylabel("Eje Y")
axes.set_title("Titulo loco")
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0.4, 0.1, 0.9, 0.8])
axes_2 = fig.add_axes([0.5, 0.6, 0.3, 0.2]) # izq, piso(abajo), ancho, altura

# 1
axes_1.plot(x, y, "r")
axes_1.set_xlabel("Tiempo")
axes_1.set_ylabel("Distancia")
axes_1.set_title("Carrera")

# 2 
axes_2.plot(y, x, "b")
axes_2.set_xlabel("Tiempo Juan")
axes_2.set_ylabel("Distancia Juan")
axes_2.set_title("Carrera Juan") 
```

## OO Subplots

```python
fig, axes = plt.subplots(nrows = 3, ncols = 4)
plt.tight_layout()
```

```python
count = 0
for p in axes: 
    for i in p: 
        if count % 2 == 0:
            i.plot(x, y, 'r')
        else:
            i.plot(y, x, 'y')
        count += 1

plt.tight_layout() 
fig 
```

```python
fig, axes = plt.subplots(figsize=(12,5), dpi=150)

axes.plot(x, y, "b*")
axes.set_xlabel("Resolucion")
axes.set_ylabel("Velocidad Refresh")
axes.set_title("Video") 
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, "r", label="x^2")
ax.plot(x, x**3, "b", label="x^3")
ax.plot(x, x**4, "g", label="x^4")
ax.legend(loc=7) # "best" 0 
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, "r+--", label="x^2")
ax.plot(x, x**3, "b*--", label="x^3")
ax.plot(x, x**4, "g.--", label="x^4")
ax.legend(loc=10) # "center" 
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.2)
ax.plot(x, x+3, color="#FF5733")
ax.set_facecolor("#f6e0b7") 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‚Äò-‚Äò, ‚Äò‚Äì‚Äô, ‚Äò-.‚Äô, ‚Äò:‚Äô, ‚Äòsteps‚Äô
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
```
