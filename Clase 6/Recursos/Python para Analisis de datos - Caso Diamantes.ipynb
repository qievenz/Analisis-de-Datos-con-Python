## Boxplot

También conocido como diagrama de caja y bigote. Es un método estandarizado para representar gráficamente una serie de datos numéricos a través de sus cuartiles. De esta manera, el diagrama de caja muestra a simple vista la mediana y los cuartiles de los datos, pudiendo también representar los valores atípicos de estos(outliers).

¿Recordas que decía que el histograma es muy útil para observar cómo se distribuye tu variable numérica? El Boxplot también lo es. Si lo miras con mas detalle, incluso te revelará más información en menos tiempo.

El máximo interés del box-plot es visualizar la distribución de una variable numérica de la manera más simplificada posible. Sólo utiliza los valores de los cuartiles, los extremos (el mínimo y el máximo) y valores raros o outliers.

No depende de valores ponderados como la media. Simplemente se fija en las características de la posición. En el fondo, primero ordena la variable y después la caracteriza según la posición de los cuartiles y los valores extremos.

La gracia del boxplot es visualizar una variable numérica de diferentes grupos o categorías. Te facilita la vida cuando comparas varios grupos.

```python
# BOX Carat / Claridad
diamonds.boxplot(column="carat", by ="clarity", figsize=(8,8), notch = True)
plt.show()

# notch nos permite hacer una muesca en la mediana para marcar donde se ubica dicho valor.
```

La mayoria de los diamantes se situan en la parte inferior de la escala
Muchos outliers (mas que 3.5, ya que el maximo 1.5 del rango intercuartilico es cerca de 2 y a partir de ahi toma como outliers

```python
# Precio / Cut
green_diamond = dict(markerfacecolor="g", marker="D")
diamonds.boxplot(column="price", by ="cut", figsize=(8,8), flierprops=green_diamond )
plt.show()

# Creamos un diccionario con las propiedades de los outliers para configurarlos con el color y el tipo de marcador
```

```python
# Precio / Color
diamonds.boxplot(column="price", by ="color", figsize=(8,8), showfliers=False, vert=False)
plt.show()

# Boxplot sin outliers
# Boxplot horizontal
```

## Densidades

## Tabla de frecuencias y barplot

```python
carat_table = pd.crosstab(index=diamonds['clarity'],
                          columns='count')
print(carat_table)
```

```python
carat_table.plot(kind="bar", figsize=(8,8), alpha=0.3, grid=True)
plt.show()
```

```python
carat_table2=pd.crosstab(index=diamonds["clarity"], columns=diamonds['color'])
print(carat_table2)
carat_table2.plot(kind="bar", figsize=(8,8), alpha=0.9, grid=True, stacked=True)
plt.show()
```

```python
carat_table2=pd.crosstab(index=diamonds["color"], columns=diamonds['cut'])
print(carat_table2)
carat_table2.plot(kind="bar", figsize=(8,8), alpha=0.9, grid=True, stacked=False)
plt.show()
```
