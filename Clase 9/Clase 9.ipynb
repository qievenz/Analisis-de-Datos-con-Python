```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Parte 1

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,5)) 
ax_1 = axes[0] 
ax_1.plot(x, y, color="blue", lw=3, ls="-") 
ax_1.set_xlabel("x") 
ax_1.set_ylabel("y") 
ax_1.set_title("y(x)") 

ax_2 = axes[1] 
ax_2.plot(x, z, color="red", lw=5, ls=":") 
ax_2.set_xlabel("x") 
ax_2.set_ylabel("z") 
ax_2.set_title("z(x)") 

plt.tight_layout(pad=3) 
```
